```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import math
import re
```

```python
address_data = pd.read_csv('AddressTrip.csv')
```

```python
address_data.head()
```

Xóa những dòng dữ liệu null

```python
address_data_dropna=address_data.dropna()
address_data = address_data_dropna.reset_index(drop=True)
address_data.info()
```

```python
address_data
```

Xóa những dòng dữ liệu có địa chỉ không rõ ràng, hoặc có kí tự đặc biệt như _!@#$%^&*~`;:'"><?


```python
address_data = address_data[~(address_data['Address'].apply(lambda x: len(x) < 30 or '_' in x))]
def is_valid_address(address):
    return re.match(r'^[a-zA-Z0-9\s,.-]*$', address) is not None

address_data = address_data[address_data['Address'].apply(is_valid_address)]
```

```python
address_data.to_csv('AddressPreprocessing.csv', index=False)
```
